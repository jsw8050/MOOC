### 2. 용어
#### 주석
- C 주석: /*     */
- C++ 주석: // 줄 단위

#### 예약어 (keyword)
- 약속된 의미의 단어
- 사용자가 임의로 재정의해서 사용할 수 없음
- union, while

#### 식별자(identifier)
- 사용자가 만든 이름 
- 첫 글자는 숫자가 오면 안된다.
- 영문 대소문자 다르게 인식

#### 상수 (constant)
- 프로그램 내에서 항상 고정된 값

#### 변수 (variable)
- 데이터를 메모리에 저장하는 공간
- 데이터를 처리하기 위해서는 데이터 타입을 이용해 변수를 선언

#### 연산자 (operator)
- 수행되는 연산을 표현하는 기호(산술, 관계, 대입, 논리, 증감 등)

#### 표현식(expression)
- 상수, 변수, 연산자, 함수 호출 드응로 구성된 의미있는 식
- 표현식은 항상 평가값을 가짐

#### 문장(statement)
- 하나의 처리 단위는 반드시 세미클론으로 끝내야 함
- 컴파일러에게 처리 단위를 알려주는 역할
- 복합문
    - 여러 문장을 묶어서 하나의 문장처럼 처리
    - 중괄호 사용
    
#### 대입문(assignment statement)
- 대입 연산자(=) 는 오른쪽에 잇는 표현식의 평가값을 왼쪽 변수에 저장

### 3. 데이터 타입
- 데이터 처리를 위해서는 데이터를 저장해야 함
- 데이터를 종류에 따라 분류하여 적절한 크기의 공간에 저장하기 위해서 데이터 타입이 필요
- 변수에 저장되는 값의 종류를 데이터 타입 또는 간단히 자료형이라고 표현
    - 기본 자료형: int, char, float
    - 유도 자료형(derived type): 배열, 구조체, 공용체
    
#### 변수 선언
- 변수: 프로그램에서 자료 값을 임시로 기억할 수 있는 저장 공간
- Ram에 저장
- 같은 데이터 타입 변수 여러 개 선언 시 콤마 사용

#### C 데이터 타입
- 정수형
    - 문자형(1byte): char, signed char(음수 기호가 있는 정수), unsigned char(모두 양수로만 표현하는 정수)
    - 정수형(2~4byte)
- 부동소수형(4~8byte): float(4), double(8), long double(8)

#### 정수형
- 큰 의미에서는 문자형 포함
- 좁은 의미에서는 순수하게 정수를 저장하는 자료형
- signed int의 범위 : 정수 자료형 signed int의 크기가 32비트(4바이트)이면 음수부터 양수까지 표현이 가능하므로 $-2^{31}$에서부터 $2^{31}-1$까지 표현 가능
- unsigned int의 범위: 0과 양수만 표현

### 4. 전처리와 입출력 함수
#### 전처리기(preprocessor)
- 전처리는 컴파일러가 프로그래밍 언어를 기계어로 바꾸기 전에 처리해야할 명령을 먼저 수행하는 것
- 다양한 라이브러리를 작업 중인 소스파일에서 사용할 수 있도록 헤더 파일을 포함 
- #include
- #define 문자열1 문자열2: 컴파일 전에 소스 파일의 문자열1을 문자열2로 치환

#### printf 함수
- 여러 개의 다양한 데이터타입을 형식화시켜 출력하는 함수
- 문자열 출력
- 변수값 출력
    - %d: 정수를 10진수로 출력
    - %f: 소수점이 있는 실수로 출력
    - %c: 문자 형태로 출력
    - %s: 문자열 형태로 출력

#### scanf 함수
- 여러개의 다양한 데이터 타입을 한 번에 입려 받을 수 있는 함수
- 키보드로부터 입력된 데이터를 지정된 형식으로 변환하여 변수에 저장
    - int a;
      scanf("%d", &a);
