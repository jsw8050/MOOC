**1) 계산 대상 바라보기**
- 알맹이 요약하기에 필요


**2) 계산 자료 바라보기-다음 유도하기**
- 작은 것으로부터 더 큰 것을 간단하게 이끌어 내기
    - 자연수 100 != 1+1+1+...+1
                 = 자연수 99 +1
    - 자연수 n = '최초 자연수 0' 또는 '직전 자연수 (n-1) 더하기 1'\

- 다음 유도하기로 바라보는 이유
    - 계산 자료가 대개 일정한 규칙에 따라 생성되기 때문
       - 작은 성적 더미 위에 + 성적 쌓기 
       - 작은 최단 거리 표에 + 최단 거리 하나 추가
       - 작은 그림 + 동그라미 선 (반복)

- 다음 유도하기의 조건
    - 작은 것의 꼴 = 큰 것의 꼴

- 다음 유도하기 이후 문제 풀기
    - 두 가지 경우로 나누어 접근
        - 최초 경우 : 0
        - 작은 것으로부터 더 큰 것을 이끌어 내는 경우: 작은 답 + 추가 -> 큰 답 (반복)
            : n-1 + 1 = n

**자연수 0부터 n까지 더하기**
- 자연수 n = '최초 자연수 0' 또는 '직전 자연수 (n-1) 더하기 1'
    - n = 0 경우:
        자연수 0 부터 0까지 합으로 0을 바로 계산 한다.
    - n = (n-1) + 1 경우 (n >= 1):
        (n-1) 부분을 처리하기 위해서 자연수 0부터 (n-1)까지 합을 구한다.
        +1 부분을 처리하기 위해서 추가로 n을 더한다.
    

**실제 계산 과정**
- 5까지 합 = 4까지 합          +5
           = 3까지 합        +4
           = 2까지 합      +3
           = 1까지 합    +2
           = 0까지 합  +1
           = 0        0
           = 1
           = 3
