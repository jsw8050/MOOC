### 5-1. 순서열 = 계산 자료의 '더미'
계산 자료 = 같은 꼴의 더미일 경우
- 계산 설계가 쉬워짐
- 알맹이 요약하기에서 중요
- 같은 꼴의 더미를 표현:[], [1], ['월요일', '화요일','수요일'] 등
- 자체로서 완전한 프로그램

#### 다음 유도하기로 더미 해석하기
- 최초 빈 더미: []
- 직전 더미에 간단하게 원소 하나를 추가해서 얻은 더미

#### 다음 유도하기에 맞추어 순서열 만들기
- 다른 기능 이용해서 흉내내기
    - 빈 순서열: []
    - 작은 순서열 맨 앞에 원소 하나를 덧붙여 얻은 순서열: [원소] + 작은 순서열
    - 예시: [1] == [1] + [], [2,3] == [2] + [3]
    
#### 순서열과 여러곱
- 여러곱
    - 원소 갯수가 고정
    - 원소 추가하면 새로운 꼴의 여러곱이 나옴
- 순서열
    - 원소가 몇 개 포함될지 미정
    - 원소를 추가해도 종류가 달라지지 않음
    
#### 순서열을 쓸 수 있는 곳
- 기본 자료가 나올 수 있는 곳 전부

#### 순서열로 표현하기
- 사람 언어: 인천/ 40km -> 파이썬 언어: ('인천', 40)
- 사람 언어: 인천/40km, 대전/163km, 강릉/237km -> [('인천', 40), ('대전', 163), ('강릉', 237)

#### 원소 두 개 여러곱 = 대응 관계
- 대응 관계 해석: ('인천', 40) -> 인천은 40km에 대응된다, 인천의 특징은 40km이다.
