### 6-1. '어떻게'
계산 = 무엇(알맹이 요약하기, 계산자료) + 어떻게(연산, 명령)

#### 연산자
- 컴퓨터 암산에 해당
- "정수 1729와 91을 더한다.": 1729(피연산자) +(연산자) 91(피연산자)
- 암산: CPU

#### 연산자 종류
- 단항 연산자: <단항 연산자><피연산자>: - 1729
- 이항 연산자: <피연산자1><연산자><피연산자2>: 1234 - 1729
- 고유의 문법이 있는 연산자

#### 파이썬 언어 연산자
- 정수 연산자
    - 이항 연산자: + - * / ** %
    - 단항 연산자: (단항)+ (단항)-
- 논리값 연산자: 논리곱and, 논리합or, 논리부정not
- 문자열 연산자: 이항 연산자 잇기 + 
    - 예: 'Hello' + ' World!' = 'Hello World!'
    - 단, 정수는 정수끼리, 문자는 문자끼리 가능
- 비교 연산자: == != < > <= >=
- 조건 연산자: if else
- 여러곱 연산자: ..... [...]
- 순서열 연산자: ....[0] ...[1:]
- 여러합 연산자: ((lambda x:....)(...[1]) if ...else....)

### 6-2. 연산식
게산자료+연산자 = 연산식 ->(프로그램실행) -> 연산값

#### 연산식 만들기
- 계산 자료 자체: <계산 자료>
- 연산식 자체가 피연산자: <단항 연산자><연산식> 

#### 연산식 실행
- 피연산자로부터 연산값 구한 뒤 연사자 처리

<연산식1><이항연산자><연산식2>

<연산값1><이항연산자><연산식2>

<연산값1><이항연산자><연산값2>

<최종 연산값>


#### 파이썬 언어 연산자 우선 순위
지수승 ** > 부호 그대로 두기 +, 부호 바꾸기 - > 곱셈, 나눗셈, 나머지 > 덧셈, 뺄셈, 문자열 잇기 > ....

#### 계산 자료 대신 연산식 쓰기
연산식 의미 = 연산값 = 계산 자료

### 비교 연산자
- 논리값 계산하는 이항 연산자
- 왼쪽으로 결합
- ==: True
- !=: True
- False < True
- 'HELLO' < 'Hello' : 대문자 E가 소문자 e보다 먼저 나오기 때문에 작음
- 'Hello' < 'Hello World!': 빈 문자열은 작음
